## Processed dataset (Encoded binary variables) 

```python
# Load necessary library and dataset

import pandas as pd

diabetes = pd.read_csv("../data/raw/diabetes_data_upload.csv")
```

```python
# Encode binary variables

diabetes_encoded = diabetes.replace({
    'Yes': 1, 'No': 0,
    'Positive': 1, 'Negative': 0,
    'Male': 1, 'Female': 0
})
```

```python
# Confirm encoding worked

diabetes_encoded.head()
```

```python
diabetes_encoded.info()
```

```python
# Cleaning variable names

diabetes_encoded.columns = (
    diabetes_encoded.columns
    .str.lower()          # make lowercase
    .str.replace(' ', '_') # replace spaces with underscores
)

print(diabetes_encoded.head())
```

```python
diabetes_encoded.rename(columns={
    'class': 'diabetes'
}, inplace=True)

print(diabetes_encoded.head())
```

```python
# Save the encoded dataset

diabetes_encoded.to_csv("diabetes_data_encoded.csv", index=False)
```
